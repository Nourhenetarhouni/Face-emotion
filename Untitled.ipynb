```python
import cv2
```

```python
from deepface import DeepFace
```

```python
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('happywomen.jpg')
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_RGB2BGR))
```

```python
predictions=DeepFace.analyze(img)
```

```python
predictions
```

```python
def get_dominant_emotion(data):
    if isinstance(data, list) and len(data) > 0:
        first_entry = data[0]
        if 'dominant_emotion' in first_entry:
            return first_entry['dominant_emotion']
    return None
```

```python
emotion = get_dominant_emotion(predictions)
```

```python
emotion 
```

```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml') 
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    #print(faceCascade.emty)
faces=faceCascade.detectMultiScale(gray,1.1,4)

 #draw rect
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_RGB2BGR))
```

```python
emotion
# Define the text, font, and position
font = cv2.FONT_HERSHEY_SIMPLEX
position = (0,50)
font_scale = 1
font_color = (0, 0, 255)  # Red color in BGR
line_thickness = 2
```

```python
cv2.putText(img, emotion, position, font, font_scale, font_color, line_thickness, cv2.LINE_AA);
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
import cv2
from deepface import DeepFace

# Open the webcam
cap = cv2.VideoCapture(0)

# Check if the webcam is opened correctly
if not cap.isOpened():
    raise IOError("Can't open webcam")

while True:
    ret, frame = cap.read()

    # Check if the frame is read successfully
    if not ret:
        print("Error reading frame")
        break

    # Analyze the frame for emotion using DeepFace
    result = DeepFace.analyze(frame, actions=['emotion'])

    # Print the result to see its content
    print(result)

    # Display the dominant emotion using putText
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame,
                result[0]['dominant_emotion'],  # Adjust the index if needed
                (50, 50),
                font, 1,
                (0, 0, 255),
                2,
                cv2.LINE_AA)

    # Display the original video with emotion text
    cv2.imshow('Original Video', frame)
    frame = None
    # Break the loop if 'q' is pressed
    if cv2.waitKey(4) & 0xFF == ord('q'):
        break


# Release the webcam and close all windows
cap.release()
cv2.destroyAllWindows()
```
